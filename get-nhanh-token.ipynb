```python
import requests

# Cấu hình các thông tin
url = "https://pos.open.nhanh.vn/v3.0/app/getaccesstoken"
app_id = "76586"  # Thay thế {{appId}} bằng ID thực tế của bạn

# Payload (Dữ liệu gửi đi trong body)
payload = {
    "accessCode": "Iq479Ook0YtrOLMaDWYiZugGE5YUTGKSTVZALGrR5Tq5NU0m9hToorzndjor5BTI",
    "secretKey": "bnHt6R0dzCqzfHZXhOFfMCEGHDp41AOh010rjFWEtipSO43IvTQOqUi7yLkwFFvSwlp4uSOBEIgvMGaLmSEJfM5A31st8xIZEmhwoZRVVanlAqegkMIjyyHhrN8X5BuE"
}

# Headers
headers = {
    'Content-Type': 'application/json'
}

# Tham số trên URL (Query parameters)
params = {
    'appId': app_id
}

try:
    # Gửi POST request
    # Tham số 'json' sẽ tự động convert dictionary sang JSON string 
    # và thêm Content-Type header phù hợp.
    # Requests mặc định hỗ trợ follow redirect (--location)
    response = requests.post(url, params=params, headers=headers, json=payload)

    # Kiểm tra mã phản hồi HTTP
    response.raise_for_status()

    # In kết quả trả về
    print("Status Code:", response.status_code)
    print("Response Body:", response.json())

except requests.exceptions.RequestException as e:
    print("Đã xảy ra lỗi:", e)
```
